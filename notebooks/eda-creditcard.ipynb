Import Necessary Modules

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta
```

```python
credit = pd.read_csv('../data/raw/creditcard.csv')
credit.head()
```

```python
credit.info()
```

Data Cleaning

```python
credit.isna().sum() # no null values
```

```python
credit.drop_duplicates(inplace=True) # drop duplicates
```

EDA

```python
# Univariate analysis

sns.histplot(data=credit, x='Amount', bins=50, kde=True)
plt.title('Distribution of Amount')
```

```python
sns.countplot(x='Class', data=credit)
plt.title('Distribution of Fraudulent vs Non-Fraudulent Transactions')
```

```python
# Bivariate analysis

plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='Amount', data=credit)
plt.title('Amount by Transaction Class')
plt.yscale('log')
plt.show()
```

```python
# class distribution analysis

class_counts = credit['Class'].value_counts(normalize=True)
print("Class Distribution:\n", class_counts)
```
